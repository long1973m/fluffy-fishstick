# 客户RFM分析

```python
print(df.columns)
```

```python
df[['客户消费频次', '客户个人消费总额','客户最近一次消费时间']].describe()
```

```python
# --------------------- 1. RFM分箱与评分 ---------------------
# R_Score：最近一次消费时间越近得分越高（数值越小，得分越高）
# 使用分位数分箱（q=5），标签反向分配为[5,4,3,2,1]
df['R_Score'] = pd.qcut(
    df['客户最近一次消费时间'],
    q=5,
    labels=[5, 4, 3, 2, 1]
)

# F_Score：消费频次越高得分越高
# 使用分位数分箱（q=5），标签正向分配为[1,2,3,4,5]
df['F_Score'] = pd.qcut(
    df['客户消费频次'],
    q=5,
    labels=[1, 2, 3, 4, 5]
)

# M_Score：消费金额越高得分越高
# 使用分位数分箱（q=5），标签正向分配为[1,2,3,4,5]
df['M_Score'] = pd.qcut(
    df['客户个人消费总额'],
    q=5,
    labels=[1, 2, 3, 4, 5]
)
```

```python
# --------------------- 2. 组合RFM得分 ---------------------
# 生成RFM字符串（例如：555表示R/F/M均为最高分）
df['RFM_Score'] = df['R_Score'].astype(str) + df['F_Score'].astype(str) + df['M_Score'].astype(str)

# 计算RFM总分（R + F + M，范围3-15分）
df['RFM_Total'] = df[['R_Score', 'F_Score', 'M_Score']].sum(axis=1)
```

```python
# --------------------- 3. 客户分群逻辑 ---------------------
def categorize_rfm(row):
    """根据RFM得分定义客户群体"""
    # 高价值客户：R/F/M均为4或5分
    if row['R_Score'] >= 4 and row['F_Score'] >= 4 and row['M_Score'] >= 4:
        return '高价值客户'
    # 流失客户：R得分低（≤2）但F/M得分高（≥3）
    elif row['R_Score'] <= 2 and (row['F_Score'] >= 3 or row['M_Score'] >= 3):
        return '流失客户'
    # 新客户：R得分高（≥4）但F/M得分低（≤2）
    elif row['R_Score'] >= 4 and (row['F_Score'] <= 2 or row['M_Score'] <= 2):
        return '新客户'
    # 潜力客户：R/F中等，但M得分高
    elif row['M_Score'] >= 4 and (3 <= row['R_Score'] <= 4) and (3 <= row['F_Score'] <= 4):
        return '潜力客户'
    # 一般客户：其他情况
    else:
        return '一般客户'
```

```python
# --------------------- 4. 结果输出 ---------------------
# 应用分群函数
df['Customer_Segment'] = df.apply(categorize_rfm, axis=1)

# 查看前5条数据的分群结果
print(df[['R_Score', 'F_Score', 'M_Score', 'RFM_Score', 'RFM_Total', 'Customer_Segment']].head())

# 统计各分群客户数量
print("\n客户分群统计：")
print(df['Customer_Segment'].value_counts())
```

```python
# 客户分群分布饼图
plt.figure(figsize=(6, 6))
df['Customer_Segment'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('客户分群分布')
plt.ylabel('')
plt.show()
```

```python
# RFM得分相关性热力图
rfm_corr = df[['R_Score', 'F_Score', 'M_Score']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(rfm_corr, annot=True, cmap='coolwarm')
plt.title('RFM得分相关性热力图')
plt.show()
```

可以看到，F_Score 与 M_Score的相关性系数为 0.7，呈现较强的正相关。说明消费频次高的客户，消费金额往往也较高。这类客户通常是企业的核心客户，对企业营收贡献较大，比如一些经常购买且每次消费金额都不低的忠实客户。

# 客户RFM分析（优化版）

```python
current_date = df['订单日期'].max()
print(current_date)
```

```python
# 假设最大日期为当前日期
current_date = df['订单日期'].max()

# 计算Recency
df['Recency'] = (current_date - df['订单日期']).dt.days

# 计算Frequency
df['Frequency'] = df.groupby('客户ID')['订单ID'].transform('count')

# 计算Monetary
df['Monetary'] = df.groupby('客户ID')['销售额'].transform('sum')
```

```python
# 假设业务中“60天内消费”的客户价值显著高于其他客户，可以手动定义分段
bins = [0, 60, 180, 360, 720, np.inf]
labels = [5, 4, 3, 2, 1]
df['R_Score'] = pd.cut(df['Recency'], bins=bins, labels=labels)
```

```python
# 假设业务中“消费频次≥10次”的客户价值显著高于其他客户，可以手动定义分段
bins = [0, 10, 50, 100, 200, np.inf]
labels = [1, 2, 3, 4, 5]
df['F_Score'] = pd.cut(df['Frequency'], bins=bins, labels=labels)
```

```python
# 假设业务中“消费金额≥10000”的客户价值显著高于其他客户，可以手动定义分段
bins = [0, 1000, 5000, 10000, 20000, np.inf]
labels = [1, 2, 3, 4, 5]
df['M_Score'] = pd.cut(df['Monetary'], bins=bins, labels=labels)
```

```python
# 生成RFM
df['RFM_Score'] = df['R_Score'].astype(str) + df['F_Score'].astype(str) + df['M_Score'].astype(str)
```

```python
# 计算RFM总分
df['RFM_Total'] = df[['R_Score', 'F_Score', 'M_Score']].sum(axis=1)
```

```python
# 将 R_Score 列从分类数据转换为字符串类型
df['R_Score'] = df['R_Score'].astype(str)

# 将 R_Score 列转换为数值类型，并用 0 填充 NaN 值
df['R_Score'] = pd.to_numeric(df['R_Score'], errors='coerce').fillna(0)

# 检查转换后的 R_Score 列的数据类型
print("\nConverted R_Score Column Type:", df['R_Score'].dtype)

# 计算 R_Score 的平均值
Avg_R_Score = df['R_Score'].mean()
print("\nAverage R_Score:", Avg_R_Score)
```

```python
df['F_Score'] = df['F_Score'].astype(str)

df['F_Score'] = pd.to_numeric(df['F_Score'], errors='coerce').fillna(0)

print("\nConverted R_Score Column Type:", df['F_Score'].dtype)

Avg_F_Score = df['F_Score'].mean()
print("\nAverage F_Score:", Avg_F_Score)
```

```python
df['M_Score'] = df['M_Score'].astype(str)

df['M_Score'] = pd.to_numeric(df['M_Score'], errors='coerce').fillna(0)

print("\nConverted R_Score Column Type:", df['M_Score'].dtype)

Avg_M_Score = df['M_Score'].mean()
print("\nAverage M_Score:", Avg_M_Score)
```

```python
#定义分群规则
def categorize_rfm(row):
    # 高价值客户(优质)：
    if row['R_Score'] >= Avg_R_Score and row['F_Score'] >= Avg_F_Score and row['M_Score'] >= Avg_M_Score:
        return '高价值客户'
    # 深耕客户（交易频次低）：
    elif row['R_Score'] >= Avg_R_Score and row['F_Score'] < Avg_F_Score and row['M_Score'] >= Avg_M_Score:
        return '深耕客户'
    # 挽回客户（近期交易低）：
    elif row['R_Score'] < Avg_R_Score and row['F_Score'] >= Avg_F_Score and row['M_Score'] >= Avg_M_Score:
        return '挽回客户'
    # 重要挽回客户（近期交易少，频次低）：
    elif row['R_Score'] < Avg_R_Score and row['F_Score'] < Avg_F_Score and row['M_Score'] >= Avg_M_Score:
        return '重要挽回客户'
    # 一般维持客户（近期交易少，交易低）：    
    elif row['R_Score'] < Avg_R_Score and row['F_Score'] >= Avg_F_Score and row['M_Score'] < Avg_M_Score:
        return '一般维持客户'
    # 高潜力客户（交易额低）：    
    elif row['R_Score'] >= Avg_R_Score and row['F_Score'] >= Avg_F_Score and row['M_Score'] < Avg_M_Score:
        return '高潜力客户'
    # 新客户（交易频次低，交易额低）：    
    elif row['R_Score'] >= Avg_R_Score and row['F_Score'] < Avg_F_Score and row['M_Score'] < Avg_M_Score:
        return '高潜力客户'       
    # 流失客户：其他情况
    else:
        return '流失客户'
```

```python
# 应用分群结果并且进行统计
df['Customer_Segment'] = df.apply(categorize_rfm, axis=1)
print(df['Customer_Segment'].value_counts())
```

```python
plt.figure(figsize=(6, 6))
df['Customer_Segment'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('客户分群分布')
plt.ylabel('')
plt.show()
```

低利润或负利润产品：Tables（桌子）利润为负，可能存在成本过高、售价过低或市场需求不足等问题，需要进一步分析原因并调整策略。Machines（机器）利润也较低，接近零，说明该产品的盈利能力较差，需要关注其经营状况。

# 市场分析

```python
market_sale = df.groupby('市场')['销售额'].sum()
market_profit = df.groupby('市场')['利润'].sum()

plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
market_sale.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('不同市场销售额占比')

plt.subplot(1,2,2)
market_profit.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('不同市场利润占比')
```

APAC是销售额和利润的核心贡献区域，占比均接近30%，且利润略高于销售额，表明其利润率可能较高。  
EU的利润占比（25.4%）显著高于销售额占比（23.2%），可能存在高利润率或成本优势。  
U.S利润占比（19.5%）略高于销售额（18.2%），而LATAM利润占比（15.1%）低于销售额（17.1%）,需关注其成本或定价策略差异。

```python
df['利润率'] = df['利润']/df['销费总额']
market_profit_percent = df.groupby('市场')['利润率'].mean()

# 绘制各市场的平均利润率条形图
plt.figure(figsize=(12, 6))
ax = sns.barplot(
    x=market_profit_percent.values,
    y=market_profit_percent.index,
    palette='coolwarm'
)

# 设置图表的标题、轴标签和网格线
ax.set_title('各个市场产品平均利润率')
ax.set_xlabel('平均利润值')
ax.set_ylabel('市场')
ax.grid(axis='x', alpha=0.75)

# 添加文本标注显示每个市场的平均利润率
for i, v in enumerate(market_profit_percent.values):
    ax.text(v + 0.01, i, f'{v:.2f}', color='black', ha='left')

plt.show()
```

APAC的产品平均利润率为0.09，与之前的销售额和利润占比一致，进一步验证其高效运营或高附加值产品的优势。  
Canada和LATAM的产品利润率分别为15%和12%，表现优秀，高于APAC，可以探究其原因。  
EMEA和Africa的产品利润率为负，表明这些市场中的产品当前处于亏损状态，需重点关注其运营问题或市场环境。
